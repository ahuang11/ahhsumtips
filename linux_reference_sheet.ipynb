# Linux Terminal Reference Sheet

### Basic Commands

"ls": list out files

"ls –l": list out files with more information including permissions, file size, and dates

"ls -a": list out all the files

"ls -1 | wc -l": count number of files and directories in current directory

"cd ANYDIRECTORY": changes directory to input directory

"cd ..": goes up one directory

"cd –": change directory into last directory

"cd": change directory back to home directory

"rm FILENAME": remove file

"rm –rf DIRECTORYNAME": delete file directory BE VERY CAREFUL WITH THIS!!!

"rmdir": deletes an empty file directory

"man COMMAND": opens instruction manual for a command

"mkdir NAMEOFDIRECTORY": makes a directory

"mv FILENAME DIRECTORYNAME": moves a file to a directory

"mv OLDFILENAME NEWFILENAME": renames a file

"cp FILENAME NEWFILENAME": copies a file

"nano FILENAME": open/create new text file

"pwd": print working directory

"touch FILENAME": create an empty file

"cat FILENAME": shows the file without actually opening it

"sudo COMMAND": utilize administrative privileges to run a command

"wget -r –np DIRECTORYURL" recursive/no parent wget, download all files downstream

"top": monitor running processes

"history": print history of what you've entered

"apt-get install PACKAGE": built-in Ubuntu installation tool

### Miscellaneous Commands

"df -h": see hard drives' capacity, availability, and usage

"tar -zcvf FILENAME DIRECTORYTOCOMPRESS": zip directory to tar file

"tar -xvf FILENAME": unzip tar file

"tar -xvzf FILENAME": unzip tar.gz file

"grep TEXT": finds matching text

"COMMAND1 | COMMAND2": Dump the output of one command to another command

"chmod ### FILENAME/DIRECTORY": change permissions

"script FILENAME": record

"COMMAND &": run command in background

"screen": open a "screen" that you can reattach (screen –r) or detach (ctrl + a, d)

"crontab -e" edit task scheduler

### GitHub Commands

"git clone URL": copy a Git repo online to local

"git status": see status of repo

"git diff": see changes before commit

"git add FILENAME": add files to be committed

"git reset FILENAME": remove/unstage accidentally added files from to be committed list

"git commit": save repo locally

"git push origin BRANCHNAME": save repo to online

"git pull origin BRANCHNAME": pull branch from server and try merging with current branch

"git checkout BRANCHNAME": switch to another branch

"git checkout -b BRANCHNAME": create new branch and switch to it

"git stash": stash uncommitted changes

### Relevant to Python Commands

"pip install PACKAGE": Python-based package installation tool

"conda install PACKAGE": Anaconda package installation tool

"conda install -c conda-forge PACKAGE": point to channel conda forge to install a package

"nano ~/.bash_profile": edit your bash profile to add aliases/shortcuts

"source ~/.bash_profile": install/activate bash profile

"jupyter notebook --port=#### --no-browser": startup jupyter notebook at input port

"ssh -N -f -L ####:localhost:#### USERNAME@IP.ADDRESS": port forward

### Exploring Data Commands

"ncdump -h FILENAME": show summary of a netCDF file

"ncdump -v VARIABLE FILENAME": show values of a variable from file

"ncdump -v -t TIMEVARIABLE FILENAME": show formatted time from file

"wgrib -v FILENAME": show inventory of a grib file

"wgrib2 GRIBFILE -netcdf NETCDFFILE": convert grib file to netcdf

"gdalinfo FILENAME": show info of gridded file

"gdallocationinfo -b 1 -wgs84 FILENAME LON LAT": transform gridded file to wgs84 and get value for a specified lon lat
